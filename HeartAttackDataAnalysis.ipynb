```python
heart_attack_df = data[data['Heart Attack Risk'] == 1]
labels = heart_attack_df["Alcohol Consumption"].value_counts().index
sizes = heart_attack_df["Alcohol Consumption"].value_counts()
#print(sizes)
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize = (6,6))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('MOST RISK BY Alcohol Consumption',color = 'red',fontsize = 15)

plt.show()
```

```python
cholesterol_bins = pd.cut(data['Cholesterol'], bins=[119, 190, 240, 400], labels=['Normal', 'Borderline', 'High'])
data['cholesterol_level'] = cholesterol_bins

plt.figure(figsize=(10, 6))
sns.countplot(x='cholesterol_level', hue='Heart Attack Risk', data=data, palette='Set1')
plt.title('Heart Attack Risk by Cholesterol Level', fontsize=16)
plt.xlabel('Cholesterol Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['cholesterol_level'], kde=True, color='skyblue', bins=30)
plt.title('Distribution of Cholesterol Levels', fontsize=16)
plt.xlabel('Cholesterol Level', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
Exercise_bins = pd.cut(data['Exercise Hours Per Week'], bins=[0, 8, 15, 20], labels=['Low','Medium','High'])
data['Exercise_level'] = Exercise_bins

plt.figure(figsize=(10, 6))
sns.countplot(x='Exercise_level', hue='Heart Attack Risk', data=data, palette='Set2')
plt.title('Heart Attack Risk by Exercise Level', fontsize=16)
plt.xlabel('Exercise Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python

age_bins = [19, 40, 65, 90]  # Define your age ranges
age_labels = ['19-40', '41-65', '66-90']
data['age_group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)
heart_attack_df = data[data['Heart Attack Risk'] == 1]

# Group by age group and count the number of heart attack cases in each group
age_group_counts = heart_attack_df['age_group'].value_counts().sort_index()
print(age_group_counts)

# Plot a bar chart for heart attack risk by age group
plt.figure(figsize=(8,6))
plt.bar(age_group_counts.index, age_group_counts.values, color='skyblue')

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Number of Heart Attack Cases')
plt.title('Heart Attack Risk by Age Group')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
